```python
#graphs
import numpy
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x = [20,150,350]
line_up, = plt.plot(x,[0.96,1.2,1.27], label='DropConnect')
line_down, = plt.plot(x,[5.86,1.04,0.89], label='DropOut')
plt.legend(handles=[line_up, line_down])
plt.title('Err vs Hidden Layers')
plt.show()
```

```python
x = [0.1,0.5,0.8]
line_up, = plt.plot(x,[1.02,0.96,1.03], label='DropConnect')
plt.legend(handles=[line_up])
plt.title('Err vs p(% of elements kept)')
plt.show()
```
